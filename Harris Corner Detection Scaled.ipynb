the scaling affects the corner detection, the corners aren't fully detected when the scaling is applied.
Harris Corner Detection is scale varient.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread(r"C:\Users\islam\Downloads\D1.png")
colored = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display the original colored image
plt.imshow(colored)
plt.title('Colored Image')
plt.axis('off')
plt.show()
```

```python
# Scale the image by 1.5
scaled_image = cv2.resize(colored, None, fx=1.5, fy=1.5)

# Convert the scaled image to grayscale
gray = cv2.cvtColor(scaled_image, cv2.COLOR_RGB2GRAY)
```

```python
# Apply Harris corner detection on the grayscale image
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
```

```python
# Dilate the result for better visualization
dst = cv2.dilate(dst, None)
```

```python
# Threshold for an optimal value
scaled_image[dst > 0.01 * dst.max()] = [0, 0, 255]
```

```python
# Display the scaled image with Harris corners
plt.imshow(scaled_image)
plt.title('Scaled Image with Harris Corners')
plt.axis('off')
plt.show()
```
